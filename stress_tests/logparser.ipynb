```python
import json
import pandas as pd
from glob import glob
```

```python
log_files = glob("logs/*.json")
log_files
```

```python
with open(log_files[0],"r") as f:
    data = json.loads(f.read())

# data
```

```python
columns = [
  'experiment',
  'max_runtime_secs',
  'max_models',
  'include_algos',
  'start_time',
  'end_time'
  ]



rows = []
for fn in log_files:
    with open(fn,"r") as f:
        data_raw = json.loads(f.read())
    
    for i in range(len(data_raw)):
      data = data_raw[i]

      row = []
      for col in columns:
        row.append(data[col])
      rows.append(row)


df = pd.DataFrame(rows, columns=columns)
df
```

```python
df.to_excel("logs_summary.xlsx")
```

```python
with open(log_files[3],"r") as f:
    data = json.loads(f.read())

print(data[0]["elapsed_time"])
pd.DataFrame.from_dict(data[0]["leaderboard"])
```

```python
df_event = pd.DataFrame.from_dict(data[0]["event_log"])

from datetime import timedelta

df_event["timestamp"] = pd.to_datetime(df_event["timestamp"]) + timedelta(hours=3)
df_event
```

```python
import numpy as np

ALGORITHMS = [
    "DRF",
    "GLM",
    "XGBoost",
    "GBM",
    "DeepLearning",
    "StackedEnsemble"
    ]


l = []
for name in df_event["name"]:
    find = np.array(ALGORITHMS)[[True if algo in str(name) else False for algo in ALGORITHMS]]
    if find: l.append(find[0])
    else: l.append(None)
df_event["algo"] = l

# [np.array(ALGORITHMS)[[True if algo in name else False for algo in ALGORITHMS]] for name in df_event["name"]]
#df_event["algo"] = [np.array(ALGORITHMS)[[True if algo in name else False for algo in ALGORITHMS]] for name in df_event["name"]]
```
